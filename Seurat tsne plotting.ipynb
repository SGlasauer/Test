```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
datadir = "/home/ucsd-train13/scratch/projects/organoids/results/processed_files/"
```

```python
tsne_df = pd.read_csv(datadir+"tsne_from_Seurat.csv", index_col=0)
tsne_df.head()
```

```python
sns.lmplot('tSNE_1','tSNE_2',tsne_df, hue = 'cluster', fit_reg=False, 
              scatter_kws={"s":15}, size=8)
plt.title("clusters")

```
