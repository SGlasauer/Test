```python
import pandas as pd
from scipy import stats
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data_dir = "/home/ucsd-train13/scratch/projects/organoids/results/processed_files/"
tpm_vg = pd.read_csv(data_dir+"tpm_all_vg_0.02.csv", index_col=0)
print(tpm_vg.shape)
tpm_vg.head()
```

```python
#zscore is standardized variation of genes.
#for every cell(row)we calculate the distance from the mean as multiple of stdev = scaling the data
stats.zscore(tpm_vg, axis=1)
```

```python
zscores = pd.DataFrame(stats.zscore(tpm_vg, axis=1), index = tpm_vg.index)
zscores.columns = tpm_vg.columns
zscores.head()
```

```python
zscores.to_csv(data_dir+"tpm_all_vg_0.02_zscores.csv")
```
